## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data
```

```python
# calculate and return the Total Purchase Value per player

SN_total=purchase_data.groupby("SN").sum()["Price"].rename("Total Purchase Value").nlargest(5)
SN_total
```

```python
# format the Total Purchase Value and create a new dataframe
top_spend=pd.DataFrame()
top_spend["Total Purchase Value"]=SN_total.map("${:,.2f}".format)
top_spend
```

```python
# calculate and return the Purchase count per player
SN_purchase=purchase_data.groupby("SN").count()["Purchase ID"].rename("Purchase Count")
SN_purchase.head()
```

```python
# merge the two data frames output the result including sort & formatting
ts_result=pd.merge(SN_purchase,top_spend,how="inner",on="SN")
ts_result["Average Purchase Price"]=SN_total/SN_purchase
ts_result=ts_result[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
ts_result = ts_result.sort_values(["Total Purchase Value"], ascending=False)
ts_result["Average Purchase Price"]=ts_result["Average Purchase Price"].map('${:,.2f}'.format)
ts_result
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data=purchase_data.groupby(["Item ID","Item Name"]).agg({'Purchase ID':'count', 'Price': 'sum'})
item_data=item_data.sort_values(by=["Price"], ascending=False)
item_data["Purchase Count"]=item_data["Purchase ID"]
item_data["Item Price"]=item_data["Price"]/item_data["Purchase Count"]
item_data["Total Purchase Value"]=item_data["Price"].map('${:,.2f}'.format)
item_data["Item Price"]=item_data["Item Price"].map('${:,.2f}'.format)
item_data2=pd.DataFrame()
item_data2["Purchase Count"]=item_data["Purchase Count"]
item_data2["Item Price"]=item_data["Item Price"]
item_data2["Total Purchase Value"]=item_data["Total Purchase Value"]
item_data2=item_data2.sort_values(by=["Purchase Count"], ascending=False)
item_data2.head(5)
```
